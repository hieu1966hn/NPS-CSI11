Bài toán: Sắp xếp là bài toán sx một danh sách dữ liệu để dễ dàng truy cập theo thứ tự tăng dần hoặc giảm dần. Tùy đặc trưng bài toán mà ta định nghĩa thế nào là tăng dần.
- Sắp xếp danh sách lớp theo tên 
- Sắp xếp file/ thư mục theo tên, ngày tạo, kích thước, ngày sửa ...
- Sắp xếp thứ tự hiển thị/ xuất hiện của các cửa hàng trên Grab, Food: Rating, Location, xú hướng...

Khi xử lý một bài toán sắp xếp, ta quan tâm đến các đặc trưng của dữ liệu cần sắp xếp: 
- Phân bổ của tập dữ liệu (ngẫu nhiên, theo thứ tự, ngược thứ tự, ..)
- Kiểu dữ liệu của các phần tử.
- Độ phức tạp

## Selection Sort
### Bài toán: Cho một dãy gồm n số nguyên, sắp xếp dãy trên theo thứ tự tăng dần.


```python
import matplotlib.pyplot as plt

def show_list(arr, title = ""):
    fig, ax = plt.subplots(figsize=(20, 4))
    ax.bar(range(len(arr)), arr)
    ax.set_title(title)
    ax.set_xlabel('index')
    ax.set_ylabel('value')
    ax.set_xticks(range(len(arr)))
    plt.show()

```

```python
number_list = [5, 9, 1, 12, 30, 35, 0, 10, 0, 2, 0, 0, 14]
show_list(number_list, title="number_list") ## Gọi hàm hiển thị biểu đồ

```

### Cách tiếp cận 
Selection Sort sử dụng cách tiếp cận đơn giản nhất:
- Tìm phần tử nhỏ nhất và chuyển lên đầu dãy
- Sau đó, tiếp tục tìm phần tử nhỏ nhất trong phần còn lại của dãy và chuyển lên vị trí thứ 2
- Tiếp tục như vậy đến khi hết dãy.

```python
def find_min(data_list, from_index):
    min_ind = from_index
    for i in range(from_index + 1, len(data_list)):
        if data_list[i] < data_list[min_ind]:
            min_ind = i
    return min_ind # vị trí giá trị nhỏ nhất

def selection_sort(data_list):
    for i in range(len(data_list) - 1):
        min_ind = find_min(data_list, i)
        if min_ind != i:
            data_list[i], data_list[min_ind] = data_list[min_ind], data_list[i]
    
```

```python
number_list_2 = number_list[:]
selection_sort(number_list_2)

print("Before sorting: {}".format(number_list))
print("Afrter sorting: {}".format(number_list_2))
show_list(number_list_2, title="number_list_2")
```

#### Nhận xét
- Thuật toán này chia danh sách cần sắp xếp thành 2 phần: đã được sắp xếp (nằm bên trái) và chưa được sắp xếp (bên phải).
- Ở mỗi lần lặp, thuật toán tìm phần tử nhỏ nhất của phần **chưa sắp xếp** để đưa vào cuối phần đã sắp xếp.

## Bubble Sort
Bài toán: Sắp xếp nổi bọt cũng xử lý bài toán như Selection Sort, nhưng với cách tiếp cận khác.
- Xét lần lượt các cặp 2 phần tử liên tiếp. Nếu phần tử đứng sau nhỏ hơn phần tử đứng trước, ta đổi chỗ 2 phần tử đó.
- Giả sử ta xét theo thứ tự index (0, 1), (1,2), (2,3), ... thì sau một lần xét qua dãy, phần tử có giá trị lớn nhất sẽ nằm ở cuối dãy.
- Như vậy, dãy cần sắp xếp cũng được chia thành 2 phần: đã được sắp xếp (bên phải) và chưa được sắp xếp (bên trái). Ta tiếp tục thực hiện việc xét các cặp ở phần chưa được sắp xếp để đưa phần tử lớn nhất vào phần đã được sắp xếp.

```python
def bubble_sort(arr):
    for i in range(len(arr) - 1, 0, -1):
        for j in range(i):
            if arr[j]> arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
        
```

```python
number_list_3 = number_list[:]
bubble_sort(number_list_3)

print("Before sorting: {}".format(number_list))
print("Afrter sorting: {}".format(number_list_3))
show_list(number_list_3, title="number_list_3")
```

## Insertion Sort
Bài toán: Sắp xếp chèn cũng là một cách tiếp cận khác của thuật toán sắp xếp

Sử dụng cách tiếp cận sau: 
- Chia danh sách thành 2 phần: phần đã được sắp xếp nằm bên trái và chưa được sắp xếp nằm bên phải.
- Giả sử ta đã sắp xếp xong i phần tử của dãy. Ta tiếp tục sắp xếp bằng cách tìm vị trí phù hợp của phần tử thứ i + 1 và **chèn** nó vào đó.
- Ta bắt đầu thuật toán bằng cách xem phần tử đã được sắp xếp là dãy gồm phần tử đầu tiên, sau đó thực hiện chèn phần tứ thứ 2, 3, 4, ... vào dãy đã sắp xếp.

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        j = i
        while j > 0 and arr[j] < arr[j-1]:
            arr[j], arr[j-1] = arr[j-1], arr[j]
            j-= 1
            
```

```python
number_list_4 = number_list[:]
insertion_sort(number_list_4)

print("Before sorting: {}".format(number_list))
print("Afrter sorting: {}".format(number_list_4))
show_list(number_list_4, title="number_list_4")
```
